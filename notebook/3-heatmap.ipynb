```python
import sys

sys.path.append("../src")
from IPython.display import display
from query_executor_heatmap import HeatmapExecutor

metadata_fpath = "/data/iharp-customized-storage/storage/post/metadata.csv"
variable = "2m_temperature"
#inbound time
start_datetime = "2020-01-01 00:00"
end_datetime = "2020-12-31 23:00"
# inbound area
max_lat = 9
min_lat = 1
min_lon = 6
max_lon = 14

tests = [
    {"start": "2020-01-01 00:00", "end": "2020-12-31 23:00", "s_res": 1, "s_agg": "min", "hm_agg":"max"},
    {"start": "2020-01-01 00:00", "end": "2021-02-10 20:00", "s_res": 1, "s_agg": "min", "hm_agg":"max"}
]
```

```python
for test in tests:
    print(f"Running test: {test}")
    qe = HeatmapExecutor(
        metadata=metadata_fpath,
        variable=variable,
        start_datetime=test["start"],
        end_datetime=test["end"],
        min_lat=min_lat,
        max_lat=max_lat,
        min_lon=min_lon,
        max_lon=max_lon,
        spatial_resolution=test["s_res"],
        spatial_aggregation=test["s_agg"],
        heatmap_aggregation_method=test["hm_agg"],
    )
    res = qe.compute()
    display(res)
```
