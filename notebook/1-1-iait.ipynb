```python
"""
1. inbound area, inbound time, year, 1: done
2. inbound area, inbound time, month, 1: done
3. inbound area, inbound time, day, 1: done
4. inbound area, inbound time, hour, 1: done
5. inbound area, inbound time, hour, 0.25: done
6. inbound area, inbound time, day, 0.25: done
7. inbound area, inbound time, month, 0.25: done
8. inbound area, inbound time, year, 0.25: done
"""

import sys

sys.path.append("../src")
from IPython.display import display
from query_executor_get_raster import GetRasterExecutor

metadata_fpath = "/data/iharp-customized-storage/storage/post/metadata.csv"
variable = "2m_temperature"
# inbound area
max_lat = 9
min_lat = 1
min_lon = 6
max_lon = 14

tests = [
    {"start": "2020-01-01 00:00", "end": "2020-12-31 23:00", "t_res": "year", "t_agg": "max", "s_res": 1, "s_agg": "min"},
    {"start": "2020-01-01 00:00", "end": "2020-12-31 23:00", "t_res": "month", "t_agg": "max", "s_res": 1, "s_agg": "min"},
    {"start": "2020-01-01 00:00", "end": "2020-12-31 23:00", "t_res": "day", "t_agg": "max", "s_res": 1, "s_agg": "min"},
    {"start": "2020-01-01 00:00", "end": "2020-12-31 23:00", "t_res": "hour", "t_agg": None, "s_res": 1, "s_agg": "min"},
    {"start": "2020-01-01 00:00", "end": "2020-12-31 23:00", "t_res": "hour", "t_agg": None, "s_res": 0.25, "s_agg": None},
    {"start": "2020-01-01 00:00", "end": "2020-12-31 23:00", "t_res": "day", "t_agg": "max", "s_res": 0.25, "s_agg": None},
    {"start": "2020-01-01 00:00", "end": "2020-12-31 23:00", "t_res": "month", "t_agg": "max", "s_res": 0.25, "s_agg": None},
    {"start": "2020-01-01 00:00", "end": "2020-12-31 23:00", "t_res": "year", "t_agg": "max", "s_res": 0.25, "s_agg": None}
]
```

```python
for test in tests:
    print(f"Running test: {test}")
    qe = GetRasterExecutor(
        metadata=metadata_fpath,
        variable=variable,
        start_datetime=test["start"],
        end_datetime=test["end"],
        min_lat=min_lat,
        max_lat=max_lat,
        min_lon=min_lon,
        max_lon=max_lon,
        temporal_resolution=test["t_res"],
        temporal_aggregation=test["t_agg"],
        spatial_resolution=test["s_res"],
        spatial_aggregation=test["s_agg"],
    )
    res = qe.compute()
    display(res)
```
