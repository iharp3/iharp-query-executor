```python
import sys

sys.path.append("../src")
from IPython.display import display
from query_executor_timeseries import TimeseriesExecutor

metadata_fpath = "/data/iharp-customized-storage/storage/post/metadata.csv"
variable = "2m_temperature"
# inbound area
max_lat = 9
min_lat = 1
min_lon = 6
max_lon = 14

tests = [
    {"start": "2020-01-01 00:00", "end": "2021-12-31 23:00", "t_res": "year", "t_agg": "max", "ts_agg": "mean"},
    {"start": "2020-01-01 00:00", "end": "2021-12-31 23:00", "t_res": "year", "t_agg": "max", "ts_agg": "min"},
    {"start": "2020-01-01 00:00", "end": "2021-12-31 23:00", "t_res": "year", "t_agg": "max", "ts_agg": "max"},
    {"start": "2020-01-01 00:00", "end": "2021-12-31 23:00", "t_res": "month", "t_agg": "max", "ts_agg": "max"},
    {"start": "2020-01-01 00:00", "end": "2021-12-31 23:00", "t_res": "day", "t_agg": "max", "ts_agg": "max"},
    {"start": "2020-01-01 00:00", "end": "2021-12-31 23:00", "t_res": "hour", "t_agg": None, "ts_agg": "max"}
]
```

```python
for test in tests:
    print(f"Running test: {test}")
    qe = TimeseriesExecutor(
        metadata=metadata_fpath,
        variable=variable,
        start_datetime=test["start"],
        end_datetime=test["end"],
        min_lat=min_lat,
        max_lat=max_lat,
        min_lon=min_lon,
        max_lon=max_lon,
        temporal_resolution=test["t_res"],
        temporal_aggregation=test["t_agg"],
        time_series_aggregation_method = test["ts_agg"],
    )
    res = qe.compute()
    display(res)
```
